```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import MDS, Isomap
from sklearn.manifold import SpectralEmbedding

# Define the mean vector
mean = np.array([2, 3])

# Define the covariance matrix
covariance = np.array([[1.5, 0.8], [0.8, 1.2]])

# Generate random samples from the Gaussian distribution
samples = np.random.multivariate_normal(mean, covariance, size=100)
# Plot the samples
x = samples[:, 0]
y = samples[:, 1]
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Gaussian Distribution')
plt.grid(True)
# Perform PCA for dimensionality reduction
pca = PCA(n_components=1)
pca_result = pca.fit_transform(samples)

# Perform MDS for dimensionality reduction
mds = MDS(n_components=1)
mds_result = mds.fit_transform(samples)

# Perform Isomap for dimensionality reduction
isomap = Isomap(n_components=1)
isomap_result = isomap.fit_transform(samples)

# Perform Laplacian Eigenmaps for dimensionality reduction
le = SpectralEmbedding(n_components=1, affinity='nearest_neighbors')
le_result = le.fit_transform(samples)

# Plot the dimensionality-reduced data
plt.figure(figsize=(10, 5))
plt.subplot(221)
plt.scatter(pca_result, np.zeros_like(pca_result))
plt.title('PCA')
plt.subplot(222)
plt.scatter(mds_result, np.zeros_like(mds_result))
plt.title('MDS')
plt.subplot(223)
plt.scatter(isomap_result, np.zeros_like(isomap_result))
plt.title('Iso Map')
plt.subplot(224)
plt.scatter(le_result, np.zeros_like(le_result))
plt.title('Laplacian Eigenmaps')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the Gaussian distributions
means = [[2, 3], [-1, -2], [4, -1]]
covariances = [[[1.5, 0.8], [0.8, 1.2]],
               [[0.5, -0.3], [-0.3, 1.0]],
               [[1.0, 0.2], [0.2, 0.8]]]
num_samples = 100

# Generate samples from the Gaussian distributions
samples = np.concatenate([np.random.multivariate_normal(mean, cov, size=num_samples) for mean, cov in zip(means, covariances)], axis=0)
# Assign colors to each distribution
colors = ['r', 'g', 'b']

# Plot the samples with different colors for each distribution
plt.scatter(samples[:, 0], samples[:, 1], c=[colors[i // num_samples] for i in range(samples.shape[0])])

# Plot the samples
#plt.scatter(samples[:, 0], samples[:, 1])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Multiple Gaussian Distributions in 2D')
plt.grid(True)
plt.show()
# Perform PCA for dimensionality reduction
pca = PCA(n_components=1)
pca_result = pca.fit_transform(samples)

# Perform MDS for dimensionality reduction
mds = MDS(n_components=1)
mds_result = mds.fit_transform(samples)

# Perform Isomap for dimensionality reduction
isomap = Isomap(n_components=1)
isomap_result = isomap.fit_transform(samples)

# Perform Laplacian Eigenmaps for dimensionality reduction
le = SpectralEmbedding(n_components=1, affinity='nearest_neighbors')
le_result = le.fit_transform(samples)

# Plot the dimensionality-reduced data
plt.figure(figsize=(10, 5))
plt.subplot(221)
plt.scatter(pca_result, np.zeros_like(pca_result),c=np.repeat(colors, num_samples))
plt.title('PCA')
plt.subplot(222)
plt.scatter(mds_result, np.zeros_like(mds_result),c=np.repeat(colors, num_samples))
plt.title('MDS')
plt.subplot(223)
plt.scatter(isomap_result, np.zeros_like(isomap_result),c=np.repeat(colors, num_samples))
plt.title('Iso Map')
plt.subplot(224)
plt.scatter(le_result, np.zeros_like(le_result),c=np.repeat(colors, num_samples))
plt.title('Laplacian Eigenmaps')
plt.tight_layout()
plt.show()
```
